### 이상치 제거(당월_매출_금액)
- “이 상권의 한 달 매출이 얼마나 될까?“

```python
# 1. IQR 계산
Q1 = merged["당월_매출_금액"].quantile(0.25)
Q3 = merged["당월_매출_금액"].quantile(0.75)
IQR = Q3 - Q1

# 2. 이상치 기준 설정
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 3. 이상치 제거된 데이터 생성
filtered = merged[(merged["당월_매출_금액"] >= lower_bound) & (merged["당월_매출_금액"] <= upper_bound)]

# 4. 줄어든 행 개수 확인
print("처리 전:", merged.shape)
print("처리 후:", filtered.shape)
```

### 결측치 제거

```python
filtered = filtered.dropna()
print(filtered.shape)
```

#### 유동인구 vs 매출

```python
sample = filtered.sample(5000, random_state=42)  # 5000개만 무작위 추출

plt.figure(figsize=(8, 5))
sns.scatterplot(
    x=sample["총_유동인구_수"],
    y=sample["당월_매출_금액"],
    s=15
)
plt.title("샘플 데이터 - 유동인구 vs 매출")
plt.xlabel("총 유동인구 수")
plt.ylabel("당월 매출 금액")
plt.show()
```

- 왼쪽 아래 ~ 중간쯤에 점이 빽빽하게 몰려 있고
- 유동인구가 많은 구간(5 ~ 8백만)은 거의 드물고,
- 오히려 유동인구는 적은데 매출은 높은 구간도 있음

▶ 패턴이 약~간 보이긴 함
- 전반적으로 유동인구가 아주 적을 땐 매출도 낮은 경우가 많음
- 다만 직선적인 관계(예: 유동인구 많을수록 매출도 많다)는 뚜렷하지 않음
